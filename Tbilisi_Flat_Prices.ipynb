```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Tbilisi.csv")
df.longitudes > 44.65
```

```python
df = df[(df.T != 0).any()]
df = df[df.longitudes > 44.68]
```

```python

df.plot(kind="scatter", x="longitudes", y="latitudes", label = "price", figsize = (16,8))

plt.legend()
```
