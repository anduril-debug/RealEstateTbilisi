```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Tbilisi.csv")

```

```python
df = df[(df.T != 0).any()]
df = df.drop("Unnamed: 0",axis=1)
```

```python
df = df[df.price < 150000]
```

```python
x = [41, 41.5, 42]


#flats under $150k
df.plot(kind="scatter", x="longitudes", y="latitudes", label = "price", figsize = (16,8),
        c = df["price"], cmap = plt.get_cmap("jet"))

plt.legend()
```
